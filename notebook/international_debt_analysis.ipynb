```python
# Table- 1 - country_table - file name - IDS_CountryMetaData

# Step - 1 - read the required csv file using pandas and convert into dataframe
ids_country_meta_data = pd.read_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_CountryMetaData\IDS_CountryMetaData.csv", encoding="latin1")
ids_country_data_frame = pd.DataFrame(ids_country_meta_data)
print(ids_country_data_frame.head())
# Step - 1.1 - shape function is used to find how many rows and columns.
print(ids_country_data_frame.shape)
# Step - 1.2 - info function is used to show full rows and columns.
print(ids_country_data_frame.info)
# Step - 1.3 - dtypes is used to show each columns data types. 
print(ids_country_data_frame.dtypes)
# Step - 1.4 - now check the null values in column and count of null value present in each columns
print(ids_country_data_frame.isnull())          # -> if null value shows True else False
print(ids_country_data_frame.isnull().sum())

# Step - 2 - now modify the column names by replace empty space " " and hypen "-" into underscore "_" 
ids_country_data_frame.columns = [col.replace(" ","_").replace("-","_") for col in ids_country_data_frame.columns]
print(ids_country_data_frame.head())

# Step - 3 - now rename column names "Code" as "Country_Code" and "Table_Name" as "Country_Name"
ids_country_data_frame = ids_country_data_frame.rename(columns={"Code":"Country_Code", "Table_Name":"Country_Name"})
print(ids_country_data_frame.head())

# Step - 4 - now select the columns which is sufficient
ids_country_data_frame = ids_country_data_frame[['Country_Code','Country_Name','Region','Income_Group','Currency_Unit','Special_Notes']]
print(ids_country_data_frame.head())
# Step - 4.1 - Sort the "Country_Code" columns in Alphabetic order
ids_country_data_frame = ids_country_data_frame.sort_values(['Country_Code'])
print(ids_country_data_frame.head())
# Step - 4.2 - now check the null values in column and count of null value present in each columns
print(ids_country_data_frame.isnull())
print(ids_country_data_frame.isnull().sum())
# Step - 4.3 - for reference, convert dataframe to csv file and view in external application
ids_country_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_CountryMetaData\IDS_CountryMetaData_1.csv")

# Step - 5 - now fill the null value for Income_Group column as "N/A" and Special_Notes column as "No Notes Available"
ids_country_data_frame['Income_Group'] = ids_country_data_frame['Income_Group'].fillna('N/A')
ids_country_data_frame['Special_Notes'] = ids_country_data_frame['Special_Notes'].fillna('No Notes Available')
print(ids_country_data_frame) 
# step - 5.1 - now check the null values in columns and count of null value present in each columns
print(ids_country_data_frame.isnull())
print(ids_country_data_frame.isnull().sum())
# Step - 5.2 - for reference, convert dataframe to csv file and view in external application
ids_country_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_CountryMetaData\IDS_CountryMetaData_2.csv")

# Step - 6 - Now need to remove unwanted records in this table, so, the unwanted records are present in "Country_Name" 
# column like which is not a country name ( East Asia & Pacific (excluding high income),Europe & Central Asia (excluding high income),
# IDA total, IDA only, Latin America & Caribbean (excluding high income), Least developed countries: UN classification, Low income,
# Lower middle income, Low & middle income, Middle income, Middle East, North Africa, Afghanistan & Pakistan (excluding high income),
# South Asia, Sub-Saharan Africa (excluding high income), Upper middle income ).
# These unwanted name matches with null values present in Region column. 
# For that remove null value records present in "Region" Column. 
# So, dropna function is used.
ids_country_data_frame = ids_country_data_frame.dropna(subset=['Region'])
print(ids_country_data_frame)
# step - 6.1 - now check the null values in columns and count of null value present in each columns. 
# At this point there is no null values in this table
print(ids_country_data_frame.isnull())
print(ids_country_data_frame.isnull().sum())
print(ids_country_data_frame.shape)
# Step - 6.2 - for reference, convert dataframe to csv file and view in external application
ids_country_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_CountryMetaData\IDS_CountryMetaData_2.csv")

# Step - 7 - remove duplicates in the table
ids_country_data_frame = ids_country_data_frame.drop_duplicates()
# step - 7.1 - check the table size
print(ids_country_data_frame.shape)
print(ids_country_data_frame.head(10))

# Step - 8 - Now reset the index value and delete that index column, change index value from 0,1,2,.. to 1,2,3,...,
# change index column name as Country_Id
ids_country_data_frame = ids_country_data_frame.reset_index(drop=True)
ids_country_data_frame.index = ids_country_data_frame.index + 1
ids_country_data_frame.index.name = 'Country_Id'
print(ids_country_data_frame.head(10))
print(ids_country_data_frame.dtypes)

# Step - 9 - final step Convert the cleaned dataframe into csv file.
ids_country_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_CountryMetaData\IDS_CountryMetaData_Cleaned.csv")
```

```python
# Table - 3 - debt_fact_table - file name - IDS_ALLCountries_Data

# Step - 1 - now read the required csv file and convert into dataframe
ids_all_countries_data = pd.read_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_ALLCountries_Data\IDS_ALLCountries_Data.csv", encoding='latin1')
ids_all_countries_data_frame = pd.DataFrame(ids_all_countries_data)
# Step - 1.1 - now check the data
print(ids_all_countries_data_frame)
# Step - 1.2 - now check the shape of the table
print(ids_all_countries_data_frame.shape)
# Step - 1.3 - check the data types
print(ids_all_countries_data_frame.dtypes)
# Step - 1.3 - now check the null values and count of null values
print(ids_all_countries_data_frame.isnull())
print(ids_all_countries_data_frame.isnull().sum())

# Step - 2 - now modify the column names by replace empty space " " and hypen "-" into underscore "_" 
ids_all_countries_data_frame.columns = [col.replace(" ","_").replace("-","_") for col in ids_all_countries_data_frame.columns]
print(ids_all_countries_data_frame.head())

# Step - 3 - now remove spaces starting and ending of the words present in Country_Code Column
ids_all_countries_data_frame['Country_Code'] = ids_all_countries_data_frame['Country_Code'].str.strip()
print(ids_all_countries_data_frame)

# Step - 4 - Now need to remove unwanted records in this table, so, the unwanted records are present in "Country_Name" 
# column like which is not a country name ( East Asia & Pacific (excluding high income),Europe & Central Asia (excluding high income),
# IDA total, IDA only, Latin America & Caribbean (excluding high income), Least developed countries: UN classification, Low income,
# Lower middle income, Low & middle income, Middle income, Middle East, North Africa, Afghanistan & Pakistan (excluding high income),
# South Asia, Sub-Saharan Africa (excluding high income), Upper middle income ).
# These unwanted name matches with codes present in Country_Code column. 
# For that, remove codes in records present in "Country_Code" Column. 
# So, dropna function is used.
unwanted_codes = ['EAP','ECA','IDX','IDA','LAC','LDC','LMY','LIC','LMC','MNA','MIC','SAS','SSA','UMC']
ids_all_countries_data_frame = ids_all_countries_data_frame[~ids_all_countries_data_frame['Country_Code'].isin(unwanted_codes)]
print(ids_all_countries_data_frame)
ids_all_countries_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_ALLCountries_Data\IDS_ALLCountries_Data_ids_data_1.csv")
# Step - 4.1 - delete null value records in Country_Code column
ids_all_countries_data_frame = ids_all_countries_data_frame.dropna(subset=['Country_Code'])
print(ids_all_countries_data_frame)
print(ids_all_countries_data_frame.shape)
# Step - 4.2 - for reference convert daataframe into csv file
ids_all_countries_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_ALLCountries_Data\IDS_ALLCountries_Data_ids_data_2.csv")

# Step - 5 - delete the duplicate records
ids_all_countries_data_frame = ids_all_countries_data_frame.drop_duplicates()
print(ids_all_countries_data_frame.shape)

# Step - 6 - now select the no of columns to change from wide format to long format,
# change column names from string to numeric,
# melt function is used to reshape the table
year_cols = ids_all_countries_data_frame.columns[6:]
ids_all_countries_data_frame[year_cols] = ids_all_countries_data_frame[year_cols].apply(pd.to_numeric, errors='coerce')
ids_all_countries_data_frame = ids_all_countries_data_frame.melt(
    id_vars=['Country_Code', 'Country_Name', 'Series_Code', 'Series_Name'],
    value_vars=year_cols,
    var_name='Year',
    value_name='Value'
)
print(ids_all_countries_data_frame.dtypes)
# Step - 6.1 - change the year data type as int
ids_all_countries_data_frame['Year'] = ids_all_countries_data_frame['Year'].astype(int)
print(ids_all_countries_data_frame.dtypes)    
ids_all_countries_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_ALLCountries_Data\IDS_ALLCountries_Data_ids_data_3.csv")

# Step - 7 - now delete unwanted null value records from the table 
ids_all_countries_data_frame = ids_all_countries_data_frame.dropna()
print(ids_all_countries_data_frame)

# Step - 8 - now sort the value of Country_Code in Alphaetic order
ids_all_countries_data_frame = ids_all_countries_data_frame.sort_values(['Country_Code'])
print(ids_all_countries_data_frame)

# Step - 9 - now delete the records having Value is '0'
ids_all_countries_data_frame = ids_all_countries_data_frame[ids_all_countries_data_frame['Value'] != 0]
print(ids_all_countries_data_frame)

# Step - 10 - now select the columns which is sufficient
ids_all_countries_data_frame = ids_all_countries_data_frame[['Country_Code', 'Series_Code','Year','Value']]
print(ids_all_countries_data_frame.head())
print(ids_all_countries_data_frame.shape)
print(ids_all_countries_data_frame.dtypes)

# Step - 11 - now delete the duplicate records
ids_all_countries_data_frame = ids_all_countries_data_frame.drop_duplicates()
print(ids_all_countries_data_frame)
print(ids_all_countries_data_frame.shape)

# step - 12 - now reset the index column, delete the index column, change the value from 0,1,2... to 1,2,3...
# rename index column as Debt_Fact_Id
ids_all_countries_data_frame = ids_all_countries_data_frame.reset_index(drop=True)
ids_all_countries_data_frame.index = ids_all_countries_data_frame.index + 1
ids_all_countries_data_frame.index.name = 'Debt_Fact_Id'

# Step - 13 - now Convert cleaned dataframe into csv file
ids_all_countries_data_frame.to_csv(r"F:\2_My_Learning\Projects\02_international_debt_analysis\IDS_ALLCountries_Data\IDS_ALLCountries_Data_Cleaned.csv")
```
